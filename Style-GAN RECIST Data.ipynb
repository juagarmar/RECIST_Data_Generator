```python
import random
import datetime

def generate_random_date(start_date, end_date):
    delta = end_date - start_date
    random_days = random.randint(0, delta.days)
    return start_date + datetime.timedelta(days=random_days)

start_date = datetime.date(2018, 1, 1)
end_date = datetime.date(2023, 12, 31)
random_date = generate_random_date(start_date, end_date)

# To generate a random date every 4 weeks Â± 5 days:
def generate_random_date_every_4_weeks(start_date, end_date):
    delta = end_date - start_date
    random_days = random.randint(0, delta.days)
    date = start_date + datetime.timedelta(days=random_days)
    while date <= end_date:
        yield date
        date += datetime.timedelta(days=random.randint(29, 34))
        
random_dates = list(generate_random_date_every_4_weeks(start_date, end_date))
```

```python
def generate_data(subject_id):
    data = []
    num_visits = random.randint(1, 7)
    visits = ["Baseline"] + ["{}-week".format(i) for i in range(4, 4*num_visits, 4)]
    # Assign a random number of target lesions (up to 5)
    num_target_lesions = random.randint(1, 5)
    # Assign a random number of non-target lesions (up to 10)
    num_non_target_lesions = random.randint(1, 10) if random.random() > 0.9 else random.randint(1, 5)
    # Assign a random number of new lesions (up to 10, but very unlikely to have more than 3)
    num_new_lesions = random.randint(1, 10) if random.random() > 0.9 else random.randint(1, 3)
    # Assign a random visit as the end of treatment (EOT)
    eot_visit = random.choice(visits) if random.random() > 0.5 else None
    # Assign a random location for the lesions
    location = random.choice(["Liver", "Lung", "Brain",'Liver','Liver', 'Lung', 'Brain', 'Kidney'
                              , 'Lymph Nodes', 'Lung', 'Lung', 'Lymph Nodes', 'Lymph Nodes',
                             'Lung', 'Lung','Lung', 'Lung','Adrenal Gland'])
    # Assign target lesion IDs
    target_lesion_ids = ["T{}".format(i) for i in range(1, num_target_lesions + 1)]
    # Assign non-target lesion IDs
    non_target_lesion_ids = ["NT{}".format(i) for i in range(1, num_non_target_lesions + 1)]
    # Assign new lesion IDs
    new_lesion_ids = ["NL{}".format(i) for i in range(1, num_new_lesions + 1)]
    # Assign measurements for each visit
    for visit in visits:
        # Assign target lesion measurements
        for lesion_id in target_lesion_ids:
            measurement = random.normalvariate(100, 25)
            data.append([subject_id, visit, "Target", lesion_id, location, measurement])
        # Assign non-target lesion IDs
        for lesion_id in non_target_lesion_ids:
            data.append([subject_id, visit, "Non-Target", lesion_id, location])
        # Assign new lesion IDs
        for lesion_id in new_lesion_ids:
            data.append([subject_id, visit, "New", lesion_id, location])
    # Create a DataFrame from the data
    df = pd.DataFrame(data, columns=["Subject ID", "Visit", "Method", "Lesion ID", "Location", "Measurement"])
    # Add a column indicating whether the visit is the EOT
    #df["EOT"] = df["Visit"] == eot_visit
    return df
```
